```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
dataset
```

```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
sb.countplot(df['Skirt'])
```

## SUPPORT

```python

#Let’s select itemsets with a minimum of 60% Support

from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.6)
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

## CONFIDENCE

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6) # associate items
```

## LIFT

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

## Recommendation
Every customer who buys skits should also be offered pants, and every customer who buys pants should also be offered skirts.

```python
import csv
csv_filename = 'Market_Basket_Optimisation.csv'
with open(csv_filename) as f:
    reader = csv.reader(f)
    data = list(reader)
```

## Support

```python
from mlxtend.frequent_patterns import apriori
apriori(df, min_support=0.1, use_colnames=True)


```

```python
frequent_itemsets=apriori(df, min_support=0.05, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

## Confidence

```python
from mlxtend.frequent_patterns import association_rules 
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3) # associate items
```
